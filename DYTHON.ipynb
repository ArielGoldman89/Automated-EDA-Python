```python
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

# Necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Automate EDA

from dython.nominal import associations
```

## Seaborn Libraries

```python
print(sns.get_dataset_names())
```

## Load Taxis Dataset

```python
df_taxis = sns.load_dataset('taxis')
```

```python
df_taxis 
```

```python
# Visualise associations
associations(df_taxis, cmap='viridis', title='Correlation Analysis', figsize=(8,8))
```

```python
# Visualise associations with marking columns for clearer interpretation
associations(df_taxis, cmap='RdYlGn', title='Correlation Analysis', mark_columns=True, figsize=(8,8))
```
