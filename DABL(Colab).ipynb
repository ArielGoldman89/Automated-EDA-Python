## Comparing categorical variable visualizations

This example showcases the four types of visualization supported for categorical variables for classification, which are â€˜countâ€™, â€˜proportionâ€™, â€˜mosaicâ€™ and â€˜sankeyâ€™.

```python
from dabl.plot import plot_classification_categorical
```

The â€˜countâ€™ plot is easiest to understand and closest to the data, as it simply provides a bar-plot of class counts per category. However, it makes it hard to make comparisons between different categories. For example, for workclass, it is hard to see the differences in proportions among the categories.

```python
plot_classification_categorical(df_adults, target_col='income', kind='count')
```

The â€˜proportionâ€™ plot on the other hand only shows the proportion, so we can see that the proportions in state-government, government, and self-employed are nearly the same. However, â€˜proportionâ€™ does not show how many samples are in each category. How much each category is actually present in the data can be very important, though.

```python
plot_classification_categorical(df_adults, target_col='income', kind='proportion')
```

The â€˜mosaicâ€™ plot shows both the class proportions within each category (on the x axis) as well as the proportion of the category in the data (on the y axis). The â€˜mosaicâ€™ plot can be a bit busy; in particular if there are many classes and many catgories, it becomes harder to interpret.

```python
plot_classification_categorical(df_adults, target_col='income', kind='mosaic')
```

The â€˜sankeyâ€™ plot is even busier, as it combines the features of the â€˜countâ€™ plot with an alluvial flow diagram of interactions. By default, only the 5 most common features are included in the sankey diagram, which can be adjusted by calling the plot_sankey function directly.

```python
plot_classification_categorical(df_adults, target_col='income', kind='sankey')
```
